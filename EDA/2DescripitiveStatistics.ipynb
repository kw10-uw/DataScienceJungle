Can you guess what will be the mean, the median and mode for this sample?
Check your guess with R functions.

```python
sample = np.random.normal(2, 0.5, 1000)
mu =np.mean(sample)
me = np.median(sample)
mo = stats.mode(sample)[0]

print(f'Descripitive statistics from this sample are:  Mean: {mu} Median: {me} Mode: {mo}')
```

Summarizing wage data

```python
from scipy.stats.mstats import winsorize
mu_wage = np.mean(NLSY["wage"])
me_wage = np.median(NLSY["wage"])
trimmed_mean = stats.trim_mean(NLSY["wage"], proportiontocut=0.1)
winsor_mean = np.mean(stats.mstats.winsorize(NLSY["wage"], limits=[0.1,0.1]))

print(f'Mean: {mu_wage}, Median: {me_wage}, Trimmed mean: {trimmed_mean}, winsor_mean: {winsor_mean}')
```
